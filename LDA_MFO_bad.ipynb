```python
from __future__ import print_function
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
```

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
import pandas as pd
data = pd.read_csv('data_users_text.csv', sep='\t')
data=data.dropna()
```

```python
import re
data['text'] = data['text'].apply((lambda x: re.sub('[.,!@0-9:()[]]','',x)))
data['text'] = data['text'].apply((lambda x: re.sub('club','',x)))
data['text'] = data['text'].apply((lambda x: re.sub(r"http\S+",'',x)))
data['text'] = data['text'].apply((lambda x: re.sub("id",'',x)))
data['text'] = data['text'].apply((lambda x: re.sub("[id%?><-|]",'',x)))
```

```python
good = data.loc[data['sentiment'] == 'Positive']
bad = data.loc[data['sentiment'] == 'Negative']
```

```python
docs_raw = bad['text']
len(docs_raw)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk.stem

from nltk.corpus import stopwords
sw = stopwords.words("russian")
vectorizer_s = CountVectorizer(stop_words=sw, max_df = 0.5, min_df = 5)
```

```python
dtm_tf = vectorizer_s.fit_transform(docs_raw)
print(dtm_tf.shape)
```

```python
tfidf_vectorizer = TfidfVectorizer(**vectorizer_s.get_params())
dtm_tfidf = tfidf_vectorizer.fit_transform(docs_raw)
print(dtm_tfidf.shape)
```

```python
# for TF DTM
lda_tf = LatentDirichletAllocation(n_topics=10, random_state=0)
lda_tf.fit(dtm_tf)
```

```python
pyLDAvis.sklearn.prepare(lda_tf, dtm_tf, vectorizer_s)
```
