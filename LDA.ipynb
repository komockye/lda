```python
from __future__ import print_function
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
```

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
import pandas as pd
data = pd.read_csv('lda_milk.csv', sep='\t')
```

```python
import re
data['text'] = data['text'].apply(lambda x: x.lower())
data['text'] = data['text'].apply((lambda x: re.sub('[.,!@0-9]','',x)))
```

```python
good = data.loc[data['polarity'] == 'Good']
bad = data.loc[data['polarity'] == 'Bad']
```

```python
docs_raw = bad['text']
len(docs_raw)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk.stem

from nltk.corpus import stopwords
sw = stopwords.words("russian")
vectorizer_s = CountVectorizer(stop_words=sw)
```

```python
dtm_tf = vectorizer_s.fit_transform(docs_raw)
print(dtm_tf.shape)
```

```python
# for TF DTM
lda_tf = LatentDirichletAllocation(n_topics=6, random_state=1)
lda_tf.fit(dtm_tf)
```

```python
pyLDAvis.sklearn.prepare(lda_tf, dtm_tf, vectorizer_s)
```
